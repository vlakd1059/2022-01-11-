### 유튜브 영상제목, 조회수 크롤링

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver=wb.Chrome()
time.sleep(0.5)
driver.get("https://www.youtube.com/c/ALTUBE%EA%B9%80%EC%84%B1%ED%98%84TV/videos")
time.sleep(0.5)
body=driver.find_element_by_css_selector("body")
for i in range(10):
    body.send_keys(Keys.END)
    time.sleep(1)
    
title=driver.find_elements_by_id("video-title")
score=driver.find_elements_by_css_selector("#metadata-line>span:nth-child(1)")
for i in title:
    print(i.text)
for i in score:
    print(i.text)
```

```python
import pandas as pd
```

```python
options = wb.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
title_list=[]
price_list=[]
driver= wb.Chrome("C:/Users/smhrd/크롤링/chromedriver.exe", options=options)
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(0.5)
for i in range(10):
    img=driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    title=driver.find_element_by_css_selector(".itemtit")
    price=driver.find_element_by_css_selector(".price_real")
    # 순수 글자 정보만 저장해주는 리스트 작업
    title_list.append(title.text)
    price_list.append(price.text)
    driver.back()
    time.sleep(1)
dic={"상품명": title_list, "가격": price_list}
gmarket=pd.DataFrame(dic)
gmarket
gmarket.to_csv("지마켓.csv", encoding="euc-kr")
# 데이터프레임 제작
# csv형식으로 저장
    
```
